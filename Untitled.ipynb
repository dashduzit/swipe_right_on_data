```python
import pandas as pd 
df = pd.read_csv('excel-to-csv-data-chart.csv')
df.head(20)
```

```python
df[
    ['census_id', 'neighborhood', 'inclusive_growth', 'median_income', 'imprisonment_rate_2020']
    ] .sort_values(
        by='inclusive_growth',
        ascending=False
    ).head(10)
```

```python
df[
    ['census_id', 'neighborhood', 'inclusive_growth', 'median_income', 'imprisonment_rate_2020']
    ] .sort_values(
        by='inclusive_growth',
        ascending=True 
    ).head(10)
```

```python
import pandas as pd 
import matplotlib.pyplot as plt

sorted_df = df.sort_values('inclusive_growth', ascending=False)

plt.figure(figsize=(14, 6))
bars = plt.bar(
    sorted_df['neighborhood'],
    sorted_df['inclusive_growth'],
    color='#2ca02c', edgecolor='black'
)
plt.xticks(rotation=75, ha='right')
plt.xlabel('Neighborhood / Census Tracts', fontsize=12)
plt.ylabel('Inclusive Growth Score (IGS)', fontsize=12)
plt.title('Inclusive Growth Score by Neighborhood (High to Low)', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

top10 = df.sort_values('imprisonment_rate_2020', ascending=False) .head(10)
plt.figure(figsize=(12,6))
bars = plt.barh(
    top10['census_id'].astype(str) + ' | ' + top10['neighborhood'],
    top10['imprisonment_rate_2020'],
    color= '#d62728', edgecolor='black'
)
plt.xlabel('Incarceration Rate (per 100,000)', fontsize=12)
plt.title('Top 10 Census Tracts in Baltimore City by Incarceration Rate (2020)', fontsize=14)
plt.gca() .invert_yaxis()
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

top10 = df.sort_values('imprisonment_rate_2020', ascending=True) .head(10)

plt.figure(figsize=(12,6))
bars = plt.barh(
    top10['census_id'].astype(str) + ' | ' + top10['neighborhood'],
    top10['imprisonment_rate_2020'],
    color='#1f77b4', edgecolor='black'
)
plt.xlabel('Incarceration Rate (per 100,000)', fontsize=12)
plt.title('Top 10 Census Tracts in Montgomery County by Incarceration Rate (2020)', fontsize=14)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
